```python
import os
from pathlib import Path
import shutil
 
src=r'C:\Users\vikra\OneDrive\Documents\AI Notes\capstone\yolov5-master-20220314T021606Z-001\yolov5-master\runs\detect\exp2\labels'

files=os.listdir(src)
label_list=[]
   
for idx,fname in enumerate(files):
   
    label_list.append(os.path.splitext(fname)[0])
        

label_list
        
```

```python
width=1024
height=1024

bb=[]
for n in idn:
    if n in label_list:
        
        path=os.path.join(src,n)
        path=path+'.txt'
        #file=open(path)
        row=[]
        bbox_temp=[]
        with open(path, 'rt') as fd:
            for line in fd:
                stripped_line = line.strip()
                splited = stripped_line.split()
                bbox_temp.append(float(splited[5]))
                bbox_temp.append(float(splited[1])*width)
                bbox_temp.append(float(splited[2])*height)
                bbox_temp.append(float(splited[3])*width)
                bbox_temp.append(float(splited[4])*height)
                
                
            row.append(bbox_temp)
            bb.append(row)
    else:
        bb.append([[]])
df = pd.DataFrame(bb,columns=['bbox'])
```

```python
df = pd.DataFrame(bb,columns=['bbox'])
```
